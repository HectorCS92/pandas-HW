```python
# Dependencies and Setup
import pandas as pd
import os
import csv
import math
import statistics
import pandas as pd
import numpy as np
```

```python
# Files to Load
school_data = "schools_complete.csv"
student_data = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_frame = pd.read_csv(school_data)

# Read School and Student Data File and store into Pandas Data Frames
student_frame = pd.read_csv(student_data)

# Combine the data into a single dataset
school_data_Total = pd.merge(student_frame, school_frame, how="left", on=["school_name", "school_name"])

# See as a DataFrame and check it out
data_frame = pd.DataFrame(school_data_Total)
data_frame.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python

# Task 1. District Summary
# Calculate the required variables
pass_math_frame = data_frame[data_frame["math_score"]>=70]
pass_math_frame["Student ID"].nunique()
percentage_pass_math = (pass_math_frame["Student ID"].nunique() / data_frame["Student ID"].nunique())*100
pass_reading_frame = data_frame[data_frame["reading_score"]>=70]
pass_reading_frame["Student ID"].nunique()
percentage_pass_reading = (pass_reading_frame["Student ID"].nunique() / data_frame["Student ID"].nunique())*100
percentage_overall_passing = (percentage_pass_math + percentage_pass_reading)/2
school_frame["budget"].sum()

# Creating a summary DataFrame using the values found
district_summary_frame = pd.DataFrame({
    "Total Schools":[data_frame["school_name"].nunique()],
    "Total Students" : [data_frame["Student ID"].nunique()],
    "Total Budget" : [school_frame["budget"].sum()],
    "Average Math Score" : [data_frame["math_score"].mean()],
    "Average Reading Score" : [data_frame["reading_score"].mean()],
    "% Passing Math":[percentage_pass_math],
    "% Passing Reading": [percentage_pass_reading],
    "% Overall Passing Rate":[percentage_overall_passing]})

# Formatting
district_summary_frame["Total Budget"] = district_summary_frame["Total Budget"].map("${:,.2f}".format)
district_summary_frame["% Passing Math"] = district_summary_frame["% Passing Math"].map("{:,.2f}%".format)
district_summary_frame["% Passing Reading"] = district_summary_frame["% Passing Reading"].map("{:,.2f}%".format)
district_summary_frame["% Overall Passing Rate"] = district_summary_frame["% Overall Passing Rate"].map("{:,.2f}%".format)

# Print District Summary
district_summary_frame
```

```python
# Task 2. School Summary
grouped_schools_frame = data_frame.groupby(["school_name","type"])
total_students = grouped_schools_frame["Student ID"].count()
budget_per_school = grouped_schools_frame["budget"].max()
budget_per_student = budget_per_school/total_students
school_average_math = grouped_schools_frame["math_score"].mean()
school_average_reading = grouped_schools_frame["reading_score"].mean()
pass_math_per_school = pass_math_frame.groupby(["school_name"])
perc_pass_math = (pass_math_per_school["Student ID"].count() / grouped_schools_frame["Student ID"].count())*100
pass_reading_per_school = pass_reading_frame.groupby(["school_name"])
perc_pass_reading = (pass_reading_per_school["Student ID"].count()/grouped_schools_frame["Student ID"].count())*100
perc_overall_passing = (perc_pass_math+perc_pass_reading)/2
school_summary = pd.DataFrame({"Total Students" : total_students, 
                               "Total School Budget" : budget_per_school,
                               "Per Student Budget" : budget_per_student,
                               "Average Math Score" : school_average_math,
                               "Average Reading Score" : school_average_reading,
                               "% Passing Math" : perc_pass_math,
                               "% Passing Reading" : perc_pass_reading,
                               "Overall Passing Rate" : perc_overall_passing})

school_summary_frame = pd.DataFrame({"Total Students" : total_students, 
                               "Total School Budget" : budget_per_school,
                               "Per Student Budget" : budget_per_student,
                               "Average Math Score" : school_average_math,
                               "Average Reading Score" : school_average_reading,
                               "% Passing Math" : perc_pass_math,
                               "% Passing Reading" : perc_pass_reading,
                               "Overall Passing Rate" : perc_overall_passing})

school_summary_frame["Total School Budget"] = school_summary_frame["Total School Budget"].map("${:,.2f}".format)
school_summary_frame["Per Student Budget"] = school_summary_frame["Per Student Budget"].map("${:,.2f}".format)
school_summary_frame["% Passing Math"] = school_summary_frame["% Passing Math"].map("{:,.2f}%".format)
school_summary_frame["% Passing Reading"] = school_summary_frame["% Passing Reading"].map("{:,.2f}%".format)
school_summary_frame["Overall Passing Rate"] = school_summary_frame["Overall Passing Rate"].map("{:,.2f}%".format)


school_summary_frame
```
